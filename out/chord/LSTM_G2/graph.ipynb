```python
import pickle
import matplotlib.pyplot as plt

# Load data from pickle files
def load_pickle(file_name):
    with open(file_name, 'rb') as file:
        data = pickle.load(file)
    return data

train_loss_list = load_pickle('./train_loss.pkl')
val_loss_list = load_pickle('./val_loss.pkl')

train_loss = []
val_loss = []

for train in train_loss_list:
    train_loss.append(train)
for val in val_loss_list:
    val_loss.append(val)
    
# Plotting
plt.figure(figsize=(10, 5))

plt.plot(train_loss, label='Training Loss')
plt.plot(val_loss, label='Val Loss')
plt.title('Comparison of Training and Val Losses')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```
